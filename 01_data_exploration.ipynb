```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
ds = load_dataset("Alwaly/Kidney_Cancer-cancer")
train_data = ds['train']

# Set format to numpy
train_data.set_format("numpy")

# Check first image
sample = train_data[0]
print(f"Label: {sample['label']}")
print(f"Image shape: {sample['image'].shape}")
print(f"Image type: {type(sample['image'])}")

# Display
plt.figure(figsize=(6, 6))
plt.imshow(sample['image'].squeeze(), cmap='gray')
plt.axis('off')
plt.title(f"Kidney CT - Label: {sample['label']}")
plt.show()
```

```python
# Data statistics
labels = [train_data[i]['label'] for i in range(len(train_data))]

from collections import Counter
label_counts = Counter(labels)

print("Class distribution:")
for label, count in label_counts.items():
    print(f"  {label}: {count}")

print(f"\nTotal images: {len(train_data)}")

# Visualize distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(label_counts.keys(), label_counts.values(), color=['green', 'red'])
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Kidney CT Dataset - Class Distribution')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
